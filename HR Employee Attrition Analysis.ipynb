# Exploratory Data Analysis (EDA)

## 1. Employee Attrition Distribution

```python
plt.figure(figsize=(8,5))

ax = sns.countplot(
    data=df,
    x="Attrition",
    hue="Attrition",
    palette=["#E76F51", "#2A9D8F"],
    legend=False
)

for i in ax.containers:
    ax.bar_label(i)

plt.title("Employee Attrition Distribution", fontsize=16, fontweight="bold")
plt.xlabel("Attrition Status")
plt.ylabel("Number of Employees")

plt.show()
```

### Business Interpretation

Tingkat attrition sebesar **16,12%** menunjukkan bahwa perusahaan memiliki peluang untuk meningkatkan strategi retensi karyawan. Analisis lebih lanjut diperlukan untuk mengidentifikasi faktor-faktor yang memengaruhi keputusan karyawan untuk keluar, seperti lembur, pendapatan, kepuasan kerja, usia, dan lama bekerja di perusahaan.

## 2. Attrition by Department

```python
plt.figure(figsize=(8,5))

ax = sns.countplot(
    data=df,
    x="Department",
    hue="Attrition",
    palette=["#2A9D8F", "#E76F51"]
)


for container in ax.containers:
    ax.bar_label(container)

plt.title("Employee Attrition by Department", fontsize=16, fontweight="bold")
plt.xlabel("Department")
plt.ylabel("Number of Employees")

plt.xticks(rotation=10)

plt.show()
```

### Business Interpretation

Departemen Sales perlu menjadi perhatian karena menunjukkan jumlah karyawan yang keluar relatif tinggi. Hal ini dapat disebabkan oleh beban kerja, target penjualan, tekanan pekerjaan, atau faktor lainnya. Analisis lanjutan diperlukan untuk mengetahui penyebab utama sehingga perusahaan dapat menyusun strategi retensi yang lebih efektif.

## 3. Attrition Rate by Department (%)

```python
attrition_rate = (
    df.groupby("Department")["Attrition"]
      .apply(lambda x: (x == "Yes").mean() * 100)
      .sort_values(ascending=False)
)

attrition_rate
```

```python
plt.figure(figsize=(8,5))

ax = sns.barplot(
    x=attrition_rate.index,
    y=attrition_rate.values,
    hue=attrition_rate.index,
    palette="viridis",
    legend=False
)

for i, value in enumerate(attrition_rate.values):
    plt.text(
        i,
        value + 0.3,
        f"{value:.2f}%",
        ha="center",
        fontsize=11,
        fontweight="bold"
    )

plt.title("Attrition Rate by Department", fontsize=16, fontweight="bold")
plt.xlabel("Department")
plt.ylabel("Attrition Rate (%)")

plt.ylim(0, max(attrition_rate.values) + 5)

plt.show()
```

### Business Interpretation

Berdasarkan tingkat attrition, Department **Sales** merupakan area yang paling memerlukan perhatian karena memiliki persentase karyawan keluar tertinggi. Hal ini dapat mengindikasikan adanya tantangan seperti tingginya tekanan kerja, target penjualan, atau tingkat kepuasan kerja yang lebih rendah dibandingkan department lain.

Meskipun Department **Research & Development** memiliki jumlah karyawan keluar paling banyak secara absolut, tingkat attrition yang lebih rendah menunjukkan bahwa kondisi retensi karyawan di department tersebut relatif lebih baik. Oleh karena itu, perusahaan sebaiknya memprioritaskan evaluasi terhadap Department Sales untuk mengidentifikasi faktor-faktor yang menyebabkan tingginya turnover.

## 4. Attrition by OverTime

```python
plt.figure(figsize=(8,5))

ax = sns.countplot(
    data=df,
    x="OverTime",
    hue="Attrition",
    palette=["#2A9D8F", "#E76F51"]
)

for container in ax.containers:
    ax.bar_label(container)

plt.title("Employee Attrition by OverTime", fontsize=16, fontweight="bold")
plt.xlabel("OverTime")
plt.ylabel("Number of Employees")

plt.show()
```

```python
overtime_rate = (
    df.groupby("OverTime")["Attrition"]
      .apply(lambda x: (x == "Yes").mean() * 100)
      .sort_values(ascending=False)
)

overtime_rate
```

```python
plt.figure(figsize=(6,5))

ax = sns.barplot(
    x=overtime_rate.index,
    y=overtime_rate.values,
    hue=overtime_rate.index,
    palette="magma",
    legend=False
)

for i, value in enumerate(overtime_rate.values):
    plt.text(
        i,
        value + 0.3,
        f"{value:.2f}%",
        ha="center",
        fontsize=11,
        fontweight="bold"
    )

plt.title("Attrition Rate by OverTime", fontsize=16, fontweight="bold")
plt.xlabel("OverTime")
plt.ylabel("Attrition Rate (%)")

plt.ylim(0, max(overtime_rate.values) + 5)

plt.show()
```

### Business Interpretation

Berdasarkan kedua grafik di atas, karyawan yang bekerja lembur (OverTime = Yes) memiliki tingkat attrition yang jauh lebih tinggi dibandingkan karyawan yang tidak bekerja lembur. Meskipun jumlah karyawan yang tidak lembur lebih banyak, persentase attrition pada kelompok yang sering lembur mencapai sekitar 30%, sedangkan pada kelompok yang tidak lembur hanya sekitar 10%.

Hal ini menunjukkan bahwa lembur merupakan salah satu faktor yang berkaitan dengan meningkatnya risiko turnover. Oleh karena itu, perusahaan perlu mengevaluasi kebijakan lembur, menyeimbangkan beban kerja, serta memastikan kesejahteraan karyawan agar dapat meningkatkan retensi dan mengurangi tingkat attrition.

## 5. Attrition by Job Role

```python
plt.figure(figsize=(12,6))

ax = sns.countplot(
    data=df,
    x="JobRole",
    hue="Attrition",
    palette=["#2A9D8F", "#E76F51"]
)

for container in ax.containers:
    ax.bar_label(container)

plt.title("Employee Attrition by Job Role", fontsize=18)
plt.xlabel("Job Role")
plt.ylabel("Number of Employees")
plt.xticks(rotation=30, ha='right')

plt.tight_layout()
plt.show()
```

### Attrition Rate by Job Role (%)

```python
jobrole_rate = (
    df.groupby("JobRole")["Attrition"]
      .apply(lambda x: (x == "Yes").mean() * 100)
      .sort_values(ascending=False)
)

plt.figure(figsize=(12,6))

ax = sns.barplot(
    x=jobrole_rate.index,
    y=jobrole_rate.values,
    hue=jobrole_rate.index,
    legend=False,
    palette="viridis"
)

for i, value in enumerate(jobrole_rate.values):
    plt.text(i, value + 0.5, f"{value:.2f}%", ha="center")

plt.title("Attrition Rate by Job Role")
plt.xlabel("Job Role")
plt.ylabel("Attrition Rate (%)")
plt.xticks(rotation=30, ha="right")

plt.tight_layout()
plt.show()
```

### Business Interpretation

Berdasarkan kedua grafik di atas, tingkat attrition berbeda pada setiap Job Role. Meskipun beberapa posisi memiliki jumlah karyawan keluar yang tinggi karena jumlah karyawannya lebih banyak, persentase attrition menunjukkan risiko turnover yang sebenarnya pada masing-masing posisi.

Sales Representative memiliki tingkat attrition tertinggi, diikuti oleh Laboratory Technician dan Human Resources. Hal ini menunjukkan bahwa ketiga posisi tersebut memiliki risiko turnover yang lebih besar dibandingkan Job Role lainnya. Oleh karena itu, perusahaan perlu memprioritaskan evaluasi terhadap posisi-posisi tersebut dengan meninjau beban kerja, peluang pengembangan karier, sistem kompensasi, serta tingkat kepuasan kerja agar retensi karyawan dapat ditingkatkan.

## 5. Monthly Income vs Attrition

```python
plt.figure(figsize=(8,6))

sns.boxplot(
    data=df,
    x="Attrition",
    y="MonthlyIncome",
    palette=["#2A9D8F", "#E76F51"]
)

plt.title("Monthly Income by Attrition", fontsize=16, weight="bold")
plt.xlabel("Attrition")
plt.ylabel("Monthly Income")

plt.tight_layout()
plt.show()
```

```python
income_avg = (
    df.groupby("Attrition")["MonthlyIncome"]
      .mean()
      .reset_index()
)

plt.figure(figsize=(6,5))

ax = sns.barplot(
    data=income_avg,
    x="Attrition",
    y="MonthlyIncome",
    hue="Attrition",
    legend=False,
    palette=["#2A9D8F", "#E76F51"]
)

for container in ax.containers:
    ax.bar_label(container, fmt="%.0f")

plt.title("Average Monthly Income by Attrition")
plt.xlabel("Attrition")
plt.ylabel("Average Monthly Income")

plt.tight_layout()
plt.show()
```

### Business Interpretation

Berdasarkan distribusi dan rata-rata Monthly Income, karyawan yang mengalami attrition cenderung memiliki pendapatan yang lebih rendah dibandingkan karyawan yang tetap bertahan.

Hal ini mengindikasikan bahwa kompensasi kemungkinan menjadi salah satu faktor yang memengaruhi keputusan karyawan untuk meninggalkan perusahaan. Meskipun demikian, pendapatan bukanlah satu-satunya faktor sehingga perlu dianalisis bersama variabel lain seperti lembur, masa kerja, dan karakteristik pekerjaan.

Perusahaan dapat melakukan evaluasi terhadap struktur kompensasi, program penghargaan, serta peluang peningkatan pendapatan agar tetap kompetitif dan membantu meningkatkan retensi karyawan.

## 6. Age vs Attrition

```python
plt.figure(figsize=(8,6))

sns.boxplot(
    data=df,
    x="Attrition",
    y="Age",
    palette=["#2A9D8F", "#E76F51"]
)

plt.title("Age Distribution by Attrition", fontsize=16, weight="bold")
plt.xlabel("Attrition")
plt.ylabel("Age")

plt.tight_layout()
plt.show()
```

```python
age_avg = (
    df.groupby("Attrition")["Age"]
      .mean()
      .reset_index()
)

plt.figure(figsize=(6,5))

ax = sns.barplot(
    data=age_avg,
    x="Attrition",
    y="Age",
    hue="Attrition",
    legend=False,
    palette=["#2A9D8F", "#E76F51"]
)

for container in ax.containers:
    ax.bar_label(container, fmt="%.1f")

plt.title("Average Age by Attrition")
plt.xlabel("Attrition")
plt.ylabel("Average Age")

plt.tight_layout()
plt.show()
```

### Business Interpretation

Berdasarkan distribusi usia, karyawan yang mengalami attrition cenderung memiliki usia yang lebih muda dibandingkan karyawan yang tetap bekerja.

Rata-rata usia karyawan yang resign sekitar 33,6 tahun, sedangkan karyawan yang bertahan memiliki rata-rata usia sekitar 37,6 tahun. Hal ini menunjukkan bahwa karyawan pada tahap awal hingga pertengahan karier memiliki kecenderungan lebih tinggi untuk berpindah pekerjaan.

Perusahaan dapat meningkatkan retensi pada kelompok usia yang lebih muda melalui program pengembangan karier, mentoring, peningkatan kesempatan belajar, serta jalur promosi yang lebih jelas agar mereka memiliki alasan yang lebih kuat untuk bertahan di perusahaan.

## 7. Years At Company vs Attrition

```python
plt.figure(figsize=(8,5))

ax = sns.boxplot(
    data=df,
    x="Attrition",
    y="YearsAtCompany",
    palette=["#E76F51", "#2A9D8F"]
)

plt.title("Years at Company Distribution by Attrition")
plt.xlabel("Attrition")
plt.ylabel("Years at Company")

plt.show()
```

```python
years_avg = (
    df.groupby("Attrition")["YearsAtCompany"]
      .mean()
      .reset_index()
)

plt.figure(figsize=(6,5))

ax = sns.barplot(
    data=years_avg,
    x="Attrition",
    y="YearsAtCompany",
    palette=["#2A9D8F", "#E76F51"]
)

for i, v in enumerate(years_avg["YearsAtCompany"]):
    plt.text(i, v + 0.2, f"{v:.1f}", ha="center", fontsize=11)

plt.title("Average Years at Company by Attrition")
plt.xlabel("Attrition")
plt.ylabel("Average Years at Company")

plt.show()
```

### Business Interpretation

Berdasarkan lama bekerja, karyawan yang mengalami attrition umumnya memiliki masa kerja yang lebih singkat dibandingkan karyawan yang tetap bertahan.

Rata-rata lama bekerja karyawan yang resign lebih rendah daripada karyawan yang masih bekerja, menunjukkan bahwa risiko turnover cenderung lebih tinggi pada masa awal hingga pertengahan karier di perusahaan.

Perusahaan dapat meningkatkan retensi dengan memperkuat proses onboarding, program mentoring, pengembangan karier, serta evaluasi kepuasan kerja pada beberapa tahun pertama masa kerja agar karyawan lebih termotivasi untuk bertahan dalam jangka panjang.

## 8. Correlation Heatmap

```python
plt.figure(figsize=(14,10))

corr = df.select_dtypes(include="number").corr()

sns.heatmap(
    corr,
    cmap="coolwarm",
    center=0,
    annot=False,
    linewidths=0.5
)

plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

### Business Interpretation

Correlation heatmap menunjukkan hubungan antar variabel numerik dalam dataset. Sebagian besar variabel memiliki korelasi yang rendah hingga sedang, yang mengindikasikan bahwa attrition dipengaruhi oleh kombinasi beberapa faktor, bukan oleh satu variabel saja.

Beberapa variabel yang berkaitan dengan pengalaman kerja, seperti YearsAtCompany, TotalWorkingYears, JobLevel, dan MonthlyIncome, menunjukkan hubungan positif karena cenderung meningkat seiring bertambahnya masa kerja dan jenjang karier.

Hasil ini memperkuat bahwa analisis attrition perlu mempertimbangkan berbagai aspek, seperti kompensasi, pengalaman kerja, lembur, serta karakteristik pekerjaan secara bersamaan untuk memperoleh pemahaman yang lebih komprehensif.

# 9. Executive Summary

Berdasarkan hasil Exploratory Data Analysis (EDA), tingkat attrition pada perusahaan berada pada kisaran 16%, yang menunjukkan bahwa sebagian besar karyawan masih bertahan, namun terdapat kelompok tertentu yang memiliki risiko turnover lebih tinggi.

Analisis menunjukkan bahwa karyawan yang sering bekerja lembur (OverTime) memiliki tingkat attrition yang jauh lebih tinggi dibandingkan karyawan yang tidak lembur. Selain itu, beberapa department dan job role juga menunjukkan tingkat turnover yang lebih tinggi sehingga memerlukan perhatian khusus dari manajemen.

Dari sisi karakteristik karyawan, karyawan yang resign cenderung memiliki pendapatan bulanan lebih rendah, usia yang lebih muda, serta masa kerja yang lebih singkat dibandingkan karyawan yang tetap bekerja. Hasil correlation heatmap juga menunjukkan bahwa attrition tidak dipengaruhi oleh satu faktor saja, melainkan merupakan kombinasi dari beberapa aspek seperti pengalaman kerja, kompensasi, dan karakteristik pekerjaan.

Secara keseluruhan, hasil analisis mengindikasikan bahwa perusahaan perlu memfokuskan strategi retensi pada karyawan yang berada pada tahap awal karier, memiliki pendapatan lebih rendah, serta bekerja pada posisi atau kondisi kerja dengan tingkat risiko turnover yang lebih tinggi.

# 10. Business Recommendations

Berdasarkan hasil analisis, beberapa rekomendasi yang dapat dipertimbangkan perusahaan untuk mengurangi tingkat attrition adalah sebagai berikut:

1. Mengurangi beban kerja lembur yang berlebihan melalui pengelolaan jadwal kerja dan distribusi pekerjaan yang lebih seimbang.

2. Melakukan evaluasi pada department dan job role dengan tingkat attrition tertinggi untuk mengidentifikasi penyebab utama turnover.

3. Meninjau kembali struktur kompensasi dan memberikan penghargaan yang kompetitif, terutama bagi karyawan dengan pendapatan yang relatif rendah.

4. Memperkuat program onboarding, mentoring, dan pengembangan karier bagi karyawan baru agar proses adaptasi berjalan lebih baik dan meningkatkan loyalitas.

5. Menyediakan jalur pengembangan karier yang jelas melalui pelatihan, promosi, dan peningkatan kompetensi sehingga karyawan memiliki motivasi untuk bertahan lebih lama.

6. Melakukan monitoring attrition secara berkala menggunakan dashboard HR analytics sehingga perusahaan dapat mengidentifikasi risiko turnover lebih dini dan mengambil tindakan yang tepat.
